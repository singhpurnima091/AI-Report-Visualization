```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas.api.types import is_string_dtype, is_numeric_dtype 
```

```python
#Importing Dataset
data = pd.read_excel('AI Report N.xlsx') 
```

```python
#Checking the headings 
data.head() 
```

```python
# Creating a new dataframe by dropping unecessary columns
data1=data.loc[:,data.columns.drop(['#','Title','Entity 2', 'Summary', 'URL - 1'])] 
```

```python
for column in data1:
    plt.figure(column, figsize = (5,5))
    plt.title(column)
    if is_numeric_dtype(data1[column]):
        data1[column].plot(kind = 'hist')
    elif is_string_dtype(data1[column]):
        # Plotting the histograms
        data1[column].value_counts().plot(kind = 'bar')
```
